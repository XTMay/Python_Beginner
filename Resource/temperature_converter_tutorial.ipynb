## 2. 温度转换公式

### 2.1 摄氏度到华氏度

公式：**F = (C × 9/5) + 32**

- F = 华氏度
- C = 摄氏度

````python
# 手工计算几个例子
celsius_values = [0, 20, 37, 100]

print("摄氏度 → 华氏度转换表")
print("-" * 30)
````

````python
for c in celsius_values:
    f = (c * 9/5) + 32
    print(f"{c:6.1f}°C = {f:6.1f}°F")
    
print("-" * 30)
print("注：0°C = 水的冰点，100°C = 水的沸点")
````

### 2.2 温度范围和对应消息

根据作业要求，需要根据华氏温度显示不同消息：

````python
# 演示不同温度范围
test_temps = [20, 32, 50, 86, 95]

print("温度范围演示")
print("=" * 50)

for temp in test_temps:
    if temp < 32:
        message = "太冷了！(Freezing)"
    elif temp <= 86:
        message = "天气很舒适。(Pleasant)"
    else:
        message = "太热了！(Hot)"
    
    print(f"{temp}°F: {message}")
````

### 阶段 2：添加函数

将代码组织成函数，提高可读性和可重用性：

````python
# 阶段2：添加函数
print("=== 阶段2：使用函数 ===")
print()

def celsius_to_fahrenheit(celsius):
    """
    将摄氏度转换成华氏度
    
    参数:
        celsius (float): 摄氏温度
    
    返回:
        float: 华氏温度
    """
    return (celsius * 9/5) + 32
````

````python
def get_weather_message(fahrenheit):
    """
    根据华氏温度返回天气消息
    
    参数:
        fahrenheit (float): 华氏温度
    
    返回:
        str: 天气描述
    """
    if fahrenheit < 32:
        return "太冷了！"
    elif fahrenheit <= 86:
        return "天气很舒适。"
    else:
        return "太热了！"
````

### 阶段 3：添加错误处理（完整版本）

添加 try-except 块来处理无效输入：

````python
# 阶段3：完整版本（带错误处理）
print("=== 阶段3：完整版本 ===")
print()

def celsius_to_fahrenheit(celsius):
    """
    将摄氏度转换成华氏度
    
    参数:
        celsius (float): 摄氏温度
    
    返回:
        float: 华氏温度
    """
    return (celsius * 9/5) + 32

def get_weather_message(fahrenheit):
    """
    根据华氏温度返回天气消息
    
    参数:
        fahrenheit (float): 华氏温度
    
    返回:
        str: 天气描述
    """
    if fahrenheit < 32:
        return "太冷了！"
    elif fahrenheit <= 86:
        return "天气很舒适。"
    else:
        return "太热了！"

def convert_temperature(celsius_input):
    """
    主函数：转换温度并显示结果
    
    参数:
        celsius_input: 用户输入（可能是字符串或数字）
    """
    try:
        # 尝试转换成浮点数
        celsius = float(celsius_input)
        
        # 转换成华氏度
        fahrenheit = celsius_to_fahrenheit(celsius)
        
        # 获取天气消息
        message = get_weather_message(fahrenheit)
        
        # 显示结果
        print("-" * 50)
        print(f"摄氏温度：    {celsius:.2f}°C")
        print(f"华氏温度：    {fahrenheit:.2f}°F")
        print(f"天气状况：    {message}")
        print("-" * 50)
        
    except ValueError:
        print("错误：请输入有效的数字！")
        print("示例：20, 30.5, -5")
    except Exception as e:
        print(f"发生意外错误：{e}")
````

## 4. 完整解答（提交版本）

这是可以直接提交的完整代码：

````python
"""
温度转换器程序
将摄氏度转换成华氏度并提供天气描述

作者：[你的名字]
日期：[日期]
课程：CA01 - 课堂活动
"""

def celsius_to_fahrenheit(celsius):
    """
    使用公式将摄氏度转换成华氏度：F = (C * 9/5) + 32

    参数:
        celsius (float): 摄氏温度

    返回:
        float: 华氏温度
    """
    return (celsius * 9/5) + 32

def get_weather_message(fahrenheit):
    """
    根据华氏温度返回合适的天气消息

    温度范围:
    - 低于 32°F: 寒冷
    - 32°F 到 86°F: 舒适
    - 高于 86°F: 炎热

    参数:
        fahrenheit (float): 华氏温度

    返回:
        str: 天气描述消息
    """
    if fahrenheit < 32:
        return "It's freezing!"
    elif fahrenheit <= 86:
        return "The weather is pleasant."
    else:
        return "It's hot!"

def main():
    """
    主程序函数，运行温度转换器
    处理用户输入、转换和输出显示
    """
    print("=" * 50)
    print("Temperature Converter: Celsius to Fahrenheit")
    print("温度转换器：摄氏度到华氏度")
    print("=" * 50)

    try:
        # 获取用户输入并转换成浮点数
        celsius_input = input("\nEnter temperature in Celsius: ")
        celsius = float(celsius_input)

        # 转换成华氏度
        fahrenheit = celsius_to_fahrenheit(celsius)

        # 获取合适的天气消息
        message = get_weather_message(fahrenheit)

        # 格式化显示结果
        print("\n" + "-" * 50)
        print(f"Temperature in Celsius:    {celsius:.2f}°C")
        print(f"Temperature in Fahrenheit: {fahrenheit:.2f}°F")
        print(f"Weather Condition:         {message}")
        print("-" * 50)

    except ValueError:
        print("\nError: Please enter a valid numeric value.")
        print("错误：请输入有效的数字。")
        print("Examples: 20, 30.5, -5")
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")
        print(f"发生意外错误：{e}")

# 程序入口点
if __name__ == "__main__":
    main()
````

**代码说明：**

1. **文档字符串**：每个函数都有详细的说明
2. **celsius_to_fahrenheit()**：执行温度转换
3. **get_weather_message()**：根据温度返回消息
4. **main()**：主程序，处理输入输出和错误
5. **try-except**：捕获并处理错误
6. **f-strings**：格式化输出，保留2位小数
7. **if __name__ == "__main__":**：确保只在直接运行时执行

## 6. 常见错误和解决方法

### 错误 1：忘记类型转换

````python
# 错误示例
print("错误的做法：")
try:
    celsius = "20"  # 这是字符串！
    fahrenheit = (celsius * 9/5) + 32  # 会报错
except TypeError as e:
    print(f"错误：{e}")
    print("原因：不能对字符串进行数学运算\n")

# 正确示例
print("正确的做法：")
celsius = float("20")  # 转换成浮点数
fahrenheit = (celsius * 9/5) + 32
print(f"成功：{celsius}°C = {fahrenheit}°F")
````

### 错误 2：条件判断逻辑错误

````python
# 错误示例
print("错误的做法：")
fahrenheit = 50
if fahrenheit < 32:
    print("Freezing")
if fahrenheit > 86:  # 问题：中间范围没有处理！
    print("Hot")
print("没有输出任何消息 - 这是错的！\n")

# 正确示例
print("正确的做法：")
fahrenheit = 50
if fahrenheit < 32:
    print("Freezing")
elif fahrenheit <= 86:  # 使用 elif 确保覆盖所有情况
    print("Pleasant")
else:
    print("Hot")
````

### 错误 3：边界值处理不当

````python
# 测试边界值
print("测试边界值 32°F 和 86°F：\n")

# 错误做法：使用 < 而不是 <=
print("错误的做法（使用 <）：")
for temp in [32, 86]:
    if temp < 32:
        msg = "Freezing"
    elif temp < 86:  # 问题：86 不在这个范围内
        msg = "Pleasant"
    else:
        msg = "Hot"
    print(f"{temp}°F: {msg} - 86°F 应该是 Pleasant，但显示为 Hot！")

print("\n正确的做法（使用 <=）：")
for temp in [32, 86]:
    if temp < 32:
        msg = "Freezing"
    elif temp <= 86:  # 正确：包括 86
        msg = "Pleasant"
    else:
        msg = "Hot"
    print(f"{temp}°F: {msg} - 正确！")
````

### 错误 4：小数格式化错误

````python
fahrenheit = 68.33333333333

# 错误示例
print("错误的做法（没有格式化）：")
print(f"Temperature: {fahrenheit}°F")
print("问题：小数位太多\n")

# 正确示例
print("正确的做法（使用 .2f）：")
print(f"Temperature: {fahrenheit:.2f}°F")
print("完美：恰好2位小数")
````

### 错误 5：公式错误

````python
celsius = 20

# 错误示例
print("错误的做法（公式反了）：")
wrong_fahrenheit = celsius * 5 / 9 + 32  # 这是华氏到摄氏的公式！
print(f"{celsius}°C = {wrong_fahrenheit:.2f}°F")
print("错误：20°C 应该是 68°F，不是 43.11°F\n")

# 正确示例
print("正确的做法：")
correct_fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}°C = {correct_fahrenheit:.2f}°F")
print("正确！")
````

## 8. 进阶练习

### 练习 1：双向转换

````python
def fahrenheit_to_celsius(fahrenheit):
    """
    将华氏度转换成摄氏度
    公式：C = (F - 32) * 5/9
    """
    return (fahrenheit - 32) * 5/9

# 测试
print("双向转换测试：")
print("="*50)
celsius = 20
fahrenheit = celsius_to_fahrenheit(celsius)
back_to_celsius = fahrenheit_to_celsius(fahrenheit)

print(f"摄氏 → 华氏：{celsius}°C = {fahrenheit}°F")
print(f"华氏 → 摄氏：{fahrenheit}°F = {back_to_celsius}°C")
print(f"验证：{celsius} = {back_to_celsius} ？ {abs(celsius - back_to_celsius) < 0.01}")
````

### 练习 2：支持开尔文

````python
def celsius_to_kelvin(celsius):
    """
    将摄氏度转换成开尔文
    公式：K = C + 273.15
    """
    return celsius + 273.15

def kelvin_to_celsius(kelvin):
    """
    将开尔文转换成摄氏度
    公式：C = K - 273.15
    """
    return kelvin - 273.15

# 测试
print("三种温标转换：")
print("="*50)

celsius = 20
fahrenheit = celsius_to_fahrenheit(celsius)
kelvin = celsius_to_kelvin(celsius)

print(f"摄氏度：{celsius:.2f}°C")
print(f"华氏度：{fahrenheit:.2f}°F")
print(f"开尔文：{kelvin:.2f}K")

print("\n特殊温度点：")
print(f"绝对零度：{celsius_to_kelvin(-273.15):.2f}K")
print(f"水的冰点：{celsius_to_kelvin(0):.2f}K")
print(f"水的沸点：{celsius_to_kelvin(100):.2f}K")
````
